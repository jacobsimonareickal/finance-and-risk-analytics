I define a helper function as below to get dataframe name for the given dataframe object

```python
## The below helper function returns the dataframe name for the given dataframe object using the dataframe_mapping dictionary
def get_name_from_dataframe(val:pd.DataFrame, dataframe_mapping:dict) -> str:
    '''
    val: The dataframe object for which you require the name
    dataframe_mapping: Dictionary mapping with key as the dataframe name and value as the Pandas Dataframe object
    
    Returns: Dataframe Name (str)
    '''
    for key, value in dataframe_mapping.items():
        if val is value:
            return key
    return "DataFrame does not exist!!!"
```

__Now we need to check the datatypes of all the columns of all the dataframes. Of course we can check them one by one but a more faster and easier way would be to iterate over all the dataframes and check the columns and their datatypes. If the datatypes do not match, then we print an error.__

```python
## The below is the dictionary with columns as keys and expected datatype/datatypes as values

expected_datatypes = {
    'Date': 'datetime64[ns]',
    'Open': 'int64/float64',
    'High': 'int64/float64',
    'Low': 'int64/float64',
    'Close': 'int64/float64',
    'Adj Close': 'int64/float64',
    'Volume': 'int64/float64'
}

## Below I define a function that takes in a list of dataframes and checks the datatypes of all the columns. 
## It returns an error message if the datatypes does not match

def validate_columns(stocks_dataframes_list:list[pd.DataFrame], expected_datatypes:dict) -> None:
    '''
    stocks_dataframes_list: List/Array of Dataframe objects
    expected_datatypes: Expected Datatypes of the columns
    
    Returns: None
    '''
    is_valid = True
    mismatch_counter = 0
    for stock in stocks_dataframes_list:
        for col in stock.columns:
            if str(stock.dtypes[col].name) not in expected_datatypes[col]:
                print(f'Column "{col}" in Dataframe "{get_name_from_dataframe(stock, dataframe_mapping)}" does not match the expected datatype. Expected datatype is {expected_datatypes[col]} but the current datatype is {stock.dtypes[col]}')
                mismatch_counter+=1
                is_valid = False
    if is_valid:
        print('Validation successful. All columns of all dataframes have the expected datatypes.')
    else:
        print(f'There are a total of {mismatch_counter} mismatches in the datatypes of the columns. Please cast to appropriate datatypes and try again..')


## Calling the defined function to validate the columns
validate_columns(stocks_dataframes_list, expected_datatypes)
```

As we can see, the Date column in all the dataframes are of type __object__ but the expected type is __datetime__. So let's cast the Date columns and run the function again to verify.

```python
def cast_col_to_datatype(stocks_dataframes_list:list[pd.DataFrame], col:str, datatype:str) -> None:
    '''
    This helper function is used to convert a given column to the desired datatype
    
    stocks_dataframes_list: List of DataFrame objects
    col: Column name for which you want to cast
    datatype: Datatype you want to cast that input col to
    
    Return: None
    '''
    for stock in stocks_dataframes_list:
        stock[col] = stock[col].astype(datatype)
        print(f'Column {col} of DataFrame {get_name_from_dataframe(stock, dataframe_mapping)} has been cast to {datatype}')

cast_col_to_datatype(stocks_dataframes_list, 'Date', 'datetime64[ns]')
```

__Now that we have cast the Date column to date time, let's run the validate_columns function again to validate the datatypes of all the columns__

```python
validate_columns(stocks_dataframes_list, expected_datatypes)
```

### Sub-Task 2: Data Cleaning

The first task will be to check the missing values/null values in our dataframes. Let's define a function that gives us the total number of null values in each dataframe.

```python
def find_missing_values(stocks_dataframes_list:list[pd.DataFrame], dataframe_mapping=dataframe_mapping) -> list[tuple]:
    '''
    stocks_dataframes_list: List of Pandas DataFrame objects
    dataframe_mapping: A dictionary of dataframe mapping. By default, it takes the default mapping
    
    Returns: List of tuples of the dataframe name and corresponding number of missing values sorted in descending order
    of missing values
    '''
    
    temp_dict = {}
    for stock in stocks_dataframes_list:
        temp_dict[get_name_from_dataframe(stock, dataframe_mapping)] = int(stock.isnull().sum().sum())
    return sorted(temp_dict.items(), key = lambda x:x[1], reverse = True)

missing_values = find_missing_values(stocks_dataframes_list)
missing_values
```

__Now that we have got the stock dataframes and their corresponding missing values, it's still hard to work with numbers. Humans can understand graphs and colours much better. So I will plot a bar chart of dataframes and the number of missing values.__

```python
# Here I define x and y axis data to plot bar plot using seaborn
x = [i[0] for i in missing_values]
y = [i[1] for i in missing_values]

fig, ax = plt.subplots(figsize=(15, 12))
sns.barplot(x=x, y=y, ax=ax)
plt.show()
```

From the above plot, we can easily visualise the number of missing values in each of our dataframes. We also see that the last 3 dataframe [AAL, AAPL, SP500] does not have any missing values at all. Now let's proceed to handle these missing values in Pandas

There are many different ways to handle missing values. Some of the most common ways to handle missing values are:
1. Drop rows or columns that have a missing value
2. Fill with a constant value
3. Fill with an aggregated value

Here I will be going with the first one. The other two options are not preferred as we do not want to introduce variations in data by imputation

__If all values in a particular row are null, then that row will not add any information to our data. So I will be dropping rows from the dataframe if all values in that row are empty.__

```python
def drop_missing_values(stocks_dataframes_list:list[pd.DataFrame]) -> None:
    '''
    The function is used to drop all null values from the list of dataframes
    
    stocks_dataframes_list: List of dataframe objects
    
    Returns: None
    '''
    for stock in stocks_dataframes_list:
        rows = stock.shape[0]
        stock.dropna(how='all', inplace=True)
        print(f'Dropped {rows-stock.shape[0]} null rows from DataFrame "{get_name_from_dataframe(stock, dataframe_mapping)}"')

drop_missing_values(stocks_dataframes_list)
```

As we can see from the above ouput, we have dropped rows where all columns are null from our dataframes. Such rows do not add any value to our analysis. Now let's check the remaining number of null values from our dataframes by calling the function we defined earlier.

```python
missing_values = find_missing_values(stocks_dataframes_list)
missing_values
```

### Sub-Task 2: Data Merging

The entire data must be merged into a single dataframe with only the required columns. Most importantly, you must ensure that you can identify the company and the industry from the collated data for company-specific and industry analysis.

For the above subtask, I performed the following steps:
1. Filter out the required columns and discard the rest
2. Prefix company name to those columns so that we can identify the company and perform analysis easier by merging into a single dataframe
3. Merge all dataframes to a single dataframe

Note - All above steps will be implemented as functions to improve code reusability and readabilty

```python
## The columns which I will be using for my analysis
required_columns = ['Date', 'Open', 'Close', 'Volume']
exclude_columns = ['Date']

## New column mapping dictionary
def map_company_name_to_columns(stock:pd.DataFrame, exclude_columns:list):
    '''
    This is a helper function to generate dictionary mappings of old and new column names
    
    stock: DataFrame object
    exclude_columns: List of column names that should be excluded from prefixing
    
    Returns: A dictionary mapping with old column name as key and new column name as value. The new column name will
    have the company name prefixed to it. Date column will not be mapped and reatined as such
    '''
    return {col:get_name_from_dataframe(stock, final_dataframe_mapping)+'_'+col for col in stock if col not in exclude_columns}

## Function to keep only the required columns and discard the rest
def filter_columns_dataframe(stocks_dataframes_list:list[pd.DataFrame], required_columns:list) -> list[pd.DataFrame]:
    '''
    This function retains the given set of columns and discards the rest
    
    stocks_dataframes_list: List of dataframe objects
    required_columns: List of required columns that should be kept
    
    Returns: An updated list of dataframe objects
             An updated mapping dictionary of dataframe company names and their respective dataframe objects
    '''
    
    final_dataframe_mapping = {}
    final_stocks_dataframes_list = []
    for idx,stock in enumerate(stocks_dataframes_list):
        name_of_dataframe = get_name_from_dataframe(stock, dataframe_mapping)
        stock = stock[required_columns]
        final_dataframe_mapping[name_of_dataframe] = stock
        final_stocks_dataframes_list.append(stock)
    print('Step 1 (Completed): All required columns are persisted to the dataframes')
    return final_stocks_dataframes_list, final_dataframe_mapping

def prefix_company_name_to_dataframe(final_stocks_dataframes_list:list[pd.DataFrame], exclude_columns:list) -> None:
    '''
    This function prefixes the company name to all the columns except the excluded column
    
    final_stocks_dataframes_list: An updated list of dataframe objectss
    exclude_columns: List of column names that should be excluded from prefixing
    
    Returns None
    '''
    for stock in final_stocks_dataframes_list:
        new_company_mapping_dict = map_company_name_to_columns(stock, exclude_columns)
        stock.rename(new_company_mapping_dict, inplace=True, axis='columns')
    print('Step 2 (Completed): Company name has been prefixed in all the columns for each company dataframe')

## Now I will call all the defined functions in the mentioned order

# Step 1: Filter required columns from the dataframes and discard the rest
final_stocks_dataframes_list, final_dataframe_mapping = filter_columns_dataframe(stocks_dataframes_list, required_columns)

# Step 2: Prefix company name to call the columns to make datafarmes ready for merging
prefix_company_name_to_dataframe(final_stocks_dataframes_list, exclude_columns)
```

Now let's inspect the dataframes to check the renamed columns

__As you can see from the above data, the column names looks good. Now we can proceed to step 3 to merge all dataframes to a single dataframe.__

To merge the dataframes to a single dataframe, I will be using an inner join and will use 'Date' column as a key to join. Now let's define a function to accomplish this. The function will need two inputs:
1. An updated list of dataframe objects - We found this in the earlier cells and assigned the list to variable name 'final_stocks_dataframes_list'.
2. strategy - The type of join to perform

```python
def merge_all_dataframes(final_stocks_dataframes_list:list[pd.DataFrame], strategy:str) -> pd.DataFrame:
    '''
    final_stocks_dataframes_list: An updated list of dataframe objects to be merged to a single dataframe
    strategy: The type of join to perform
    
    Returns: A single merged dataframe
    '''
    left_dataframe = final_stocks_dataframes_list[0]
    for stock in final_stocks_dataframes_list[1:]:
        right_dataframe = stock
        left_dataframe = pd.merge(left=left_dataframe, right=right_dataframe, how=strategy, on='Date')
    return left_dataframe

final_stocks_df = merge_all_dataframes(final_stocks_dataframes_list, 'inner')
final_stocks_df.head(3)
```
